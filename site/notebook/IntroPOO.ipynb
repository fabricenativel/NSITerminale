## <font color=darkblue> &diams; Penser objet ! </font>

Une autre façon de **penser** notre programme est de considérer que nous allons manipuler des objets qui possèdent certaines caractéristiques et sur lesquels nous pouvons agir. 

Dans notre exemple, le dé est vu comme un objet qui possède un nombre de faces et une couleur, c'est ce que nous appellerons les <font color="navy"><b>attributs</b></font> de l'objet et que nous pouvons lancer c'est ce que nous appellerons une <font color="navy"><b>méthode</b></font> de l'objet.

Nous devons faire la différence entre le modèle d'un dé défini ci-dessus par ses attributs et méthode, c'est que nous appelerons une <font color="navy"><b>classe</b></font> et un dé en particulier que nous appelerons <font color="navy"><b>objet</b></font>.

Dans tout langage de programmation, et donc en Python, on doit disposer d'une méthode particulière permettant de créer un objet à partir de sa classe. Donnons un exemple précis, si nous indiquons qu'un dé a un nombre de faces et une couleur (nous définissions un modèle, c'est à dire une classe). Et, il faut allors  pouvoir  dire qu'on veut créer un dé en particulier à partir de ce modèle par exemple un dé orange à six faces. Créer un objet à partir de sa classe s'appelle <font color="navy"><b>instancier</b></font> la classe, on dit donc qu'un objet est l'instanciation d'une classe. Cette méthode spéciale s'appelle le <font color="navy"><b>constructeur</b></font>.

Avant de passer à l'écriture en Python et donc de découvrir la syntaxe objet de Python, résumons par un schéma notre réflexion :



## <font color=darkblue> &diams; La syntaxe de  Python </font>

* En python, la définition d'une classe commence par le mot clé `class` suivi du nom de la classe (qui par convention commence par une majuscule) et enfin du caractère `:`
* Les méthodes de la classe sont en quelque sorte des fonctions définies pour la classe et qui s'appliqueront aux objets. Par conséquent la définition d'une méthode est similaire à celle d'une fonction c'est à dire qu'on retrouve le mot clé `def`. Pour nommer l'objet auquel la méthode s'applique on utilisera le mot clé `self`
* Les méthodes spéciales en Python ont leur nom encadrés par des double underscore `__` et le constructeur en Python est `init`
Avec ces explications, vous devriez comprendre le code commenté ci-dessous :

```python
import random

# Définition d'une classe : class <nom_de_la_classe>:
class De:
    
    # Le constructeur permettant de créer des objets de cette classe
    def __init__(self,nb_faces,couleur):
        self.nombre_faces=nb_faces
        self.couleur=couleur
    
    # La méthode de lancer du dé 
    def lancer(self):
        return random.randint(1,self.nombre_faces)
```

Comme indiqué plus haut, il nous faut à présent **instancier** cette classe pour créer un objet informatique représentant un dé en particulier.

```python
un_de = De(10,"rouge")
```

Nous venons de créer *un dé* en particulier, il a 10 faces et il est rouge !

Ce dé dispose immédiatement des méthodes de classe dont il est issu, on peut donc le lancer :

```python
un_de.lancer()
```

Arrêtons nous un instant sur la syntaxe précédente, permettant d'utiliser une méthode sur un objet, elle est de la forme : `<objet>.<methode>`
Nous avons déjà utilisé à de multiples reprises cette syntaxe en Python, à titre d'exemples 

```python
L=["premier","second"]
L.append("troisième")
L
```

En effet, Python est un langage orienté objet et les listes que nous manipulons depuis longtemps sont des objets possédant une méthode `append` permettant d'ajouter une élément en fin de liste
